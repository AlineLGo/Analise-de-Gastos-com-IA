Importar a biblioteca pandas.

```python
import pandas as pd
```

#Ler os dados.

```python
df = pd.read_excel("Gastos.xls")
```

#Limpar os dados.

```python
df= df.dropna()
```

#Criar categorias automáticas (IA simples)

```python
def classificar(descricao):
  descricao = descricao.lower()
  if "mercado" in descricao:
    return "Alimentação"
  elif "restaurante" in descricao:
    return "Alimentação"
  elif "uber" in descricao:
    return "Transporte"
  elif "cinema" in descricao:
    return "Lazer"
  else:
    return "Outros"

df["categoria"] = df["Descrição"].apply(classificar)
```

#Análise

```python
total = df["Valor"].sum()
categoria_top = df.groupby("categoria")["Valor"].sum().idxmax()
```

#Insight automático

```python
if categoria_top == "Alimentação":
  insight = "Você está gastando mais com alimentação."
elif categoria_top == "Transporte":
  insight = "Seus gastos com transporte estão altos."
else:
  insight = "Seus gastos estão equilibrados"
```

#Relatório

```python
relatorio = f"""
Relatório Financeiro

Total gasto> R$ {total}
Categoria com maior gasto: {categoria_top}

Insight:
{insight}
"""
```

#Salvar

```python
with open ("relatorio_financeiro.txt", "w") as f:
  f.write(relatorio)
print(relatorio)
```
